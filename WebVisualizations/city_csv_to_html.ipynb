```python
import pandas as pd
import datetime
```

```python
# import csv to dataframe
cities = pd.read_csv("Resources/cities.csv")
cities.head()
```

```python
# reorganize the column order
cities = cities[["City_ID", "City", "Country", "Lat", "Lng", "Date","Max Temp", "Humidity", "Cloudiness", "Wind Speed"]]
cities.head()
```

```python
cities.dtypes
```

```python
cities['Date'] = pd.to_datetime(cities['Date'],unit='s')

cities.head()
                                                        
#import datetime
#timestamp = datetime.datetime.fromtimestamp(1500000000)
#print(timestamp.strftime('%Y-%m-%d %H:%M:%S'))                                                        
```

```python
cities['Date'] = cities['Date'].dt.strftime('%Y-%m-%d')
cities.head()
```

```python
cities.dtypes
```

```python
# save dataframe to html file
cities.to_html('table.html')
```
